# Fandango Scores vs. All Sites

```python
ultimate = pd.merge(fandango, all_sites, on ='FILM', how = 'inner')
ultimate.info()
```

```python
ultimate.head()
```

```python
ultimate['RT_norm'] = np.round(ultimate['RottenTomatoes'] / 20, 1)
ultimate['RT_user_norm'] = np.round(ultimate['RottenTomatoes_User'] / 20, 1)
ultimate['Metacritic_norm'] = np.round(ultimate['Metacritic'] / 20, 1)
ultimate['Metacritic_user_norm'] = np.round(ultimate['Metacritic_User'] / 2, 1)
ultimate['IMDB_norm'] = np.round(ultimate['IMDB']/ 2, 1)
```

```python
norm_df = ultimate[["STARS","RATING","RT_norm","RT_user_norm","Metacritic_norm","Metacritic_user_norm", "IMDB_norm"]]
norm_df.head()
```

# Compare Distribution of Scores Across All Sites

```python
import matplotlib.pyplot as plt
import seaborn as sns

def move_legend(ax, new_loc, **kws):
    old_legend = ax.legend_
    handles = old_legend.legendHandles
    labels = [t.get_text() for t in old_legend.get_texts()]
    title = old_legend.get_title().get_text()
    ax.legend(handles, labels, loc=new_loc, title=title, **kws)

fig, ax = plt.subplots(figsize=(12, 7), dpi=200)
sns.kdeplot(data=norm_df, clip=[0, 5], shade=True, palette='Set1', ax=ax)

move_legend(ax, new_loc='upper left')

plt.show() 
```

# Compare Fandango (STARS) & RT Critic

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data = norm_df[['STARS','RT_norm']], shade = True, clip = [0,5])
plt.title('Fandango STARS vs RT Critic Ratings')
plt.xlabel('Scoring')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.displot(data = norm_df)
plt.title('Distribution of Normalized Scores')
plt.xlabel('Scores')
plt.show()
```

# How are the worst Movies Rated Across All Platforms?

```python
sns.clustermap(data = norm_df, cmap= 'mako')
plt.title('Cluster Map for Film Ratings Across All Platforms')
plt.show()
```

Clearly Fandango is rating movies much higher than other sites, especially considering that it is then displaying a rounded up version of the rating. Let's examine the top 10 worst movies. Based off the Rotten Tomatoes Critic Ratings, what are the top 10 lowest rated movies? What are the normalized scores across all platforms for these movies? 

```python
# CODE HERE
normal = ultimate[["FILM","STARS","RATING","RT_norm","RT_user_norm","Metacritic_norm","Metacritic_user_norm", "IMDB_norm"]]
low_ten= normal.nsmallest(10,"RT_norm")
low_ten
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data = low_ten, shade = True)
plt.title('Ratings for Rotten Tomatoes 10 Worst Reviewed Films')
plt.xlabel('Ratings')
plt.show()
```
